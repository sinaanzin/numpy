```python

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'age': [23, 30, 25, 35, 28, 40, 22, 29, 33, 31],
    'gender': ['Female', 'Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Female', 'Male']
}

df = pd.DataFrame(data)
print(df)
```

```python
df['occupation'] = ['Programmer', 'Manager', 'Analyst', 'Programmer', 'Manager', 'Analyst', 'Programmer', 'Manager', 'Analyst', 'Programmer']

print(df)
```

```python
age_filtered_df = df[df['age'] >= 30]
print(age_filtered_df)
```

```python

df.to_csv('data.csv', index=False)
df_from_csv = pd.read_csv('data.csv')
print(df_from_csv)
```
